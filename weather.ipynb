```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_excel("data/weather_data.xls")
weather.head(10)
```

```python
# date time, wind intensity (categorical), precipitation (y/n)
```

```python
new_weather = weather.drop("ff3", "Nh", "H", "P","VV", "Ff", "N", "Cl", "Cm", "Ch", "RRR", "tR",
              "WW", "W1", "W2", "E", "Tg", "E'", "sss", axis=1)
```

```python
new_weather.head()
```

```python
new_weather.rename(columns={"T": "air temperature", "P": "stmospheric pressure", "U": "relative humidity",
                   "DD": "wind direction", "ff10": "maximum gust value", "Tn":"minimum air temperature",
                   "Tx":"maximum air temperature"}, inplace=True)
new_weather.head()
```

```python
new_weather.to_csv("data/weather_new.csv", index=False)
```
